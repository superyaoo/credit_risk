```python
import pandas as pd
```

```python
import lightgbm as lgb
loaded_gbm = lgb.Booster(model_file='/home/zengjunyao/to_zjy_模型训练/model_train_result/score_th_a11_20250430_50feature_ol.txt') 
```

```python
data = pd.read_pickle('/home/zengjunyao/to_zjy_模型训练/model_train_result/sample_data_with_pred_var_20250430.pkl')
```

```python
data_train = data[data['sample_type'] == 'train']
data_test = data[data['sample_type'] == 'test']
data_oot = data[data['sample_type'] == 'oot']
```

```python
import numpy as np
```

```python
def prob2score(prob):
    """
    prob转化为概率分的工具
    :param prob:
    :return:
    """
    return round(550 - 60 / np.log(2) * np.log(prob / (1 - prob)), 0)
```

```python
model_report2(data_sets = [data_train,data_test,data_oot], prob = 'predprob', target = 'label', model_obj = loaded_gbm, feature_importance_dict=None, time_col='loan_time', score_name='score_th_b7_1',
                  p2score_fun=prob2score)
```
