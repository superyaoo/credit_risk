```python
import pandas as pd
```

```python
from datetime import datetime, timedelta
```

```python
#生成的样本业务字段文件label
label = pd.read_pickle('df_sample_v11.pkl')
```

```python
label
```

```python
label['app_order_id'] = label['app_order_id'].astype(str)
```

```python
#样本的申请时间写在下面范围
start_date = datetime.strptime('2025-01-01', '%Y-%m-%d')
end_date = datetime.strptime('2025-04-04', '%Y-%m-%d')
date_list = []
current_date = start_date
#循环
while current_date <= end_date:
    date_list.append(current_date.strftime('%Y-%m-%d'))
    current_date += timedelta(days=1)
print(date_list)
```

```python
file_path = '/home/zengjunyao/model_link/ath/'
```

```python
#选取的模块名称
#model_list = ["loan_th_base_v2","loan_th_base_v3_2","sms_th_loan0_v1","app_th_bcate_v1","app_th_loan_v2","dev_th_base_v1","dev_th_base_v2","ui_th_base_v2"]
model_list = ["loan_th_base_v2",
                      "sms_th_loan0_v1",
                      "app_th_loan_v2",
                      "dev_th_base_v2","ui_th_base_v2"]
```

```python
#将选取的模块所有特征及不同时期的数据合并df_all_feature
df_all_feature = label[['app_order_id']]   #初始化df_all_feature：所有特征
# df_all_feature = pd.DataFrame()
for model in model_list:    #遍历每一个模块
    df_model = pd.DataFrame()    #每个模型创建一个 df_model
    for date in date_list:  
        #读取该模型在不同日期下的数据，将其逐次合并到 df_model里(按列拼接)
        try:
            df_tmp1 = pd.read_parquet(file_path + model + f'/df_1_{date}.parquet')
            df_tmp2 = pd.read_parquet(file_path + model + f'/df_2_{date}.parquet')
            df_model = pd.concat([df_model,df_tmp1,df_tmp2])
        except:  #同时处理可能出现的读取文件错误
            print(model,date)
    #合并：将所有模型的df_model按行合并到df_all_feature     
    df_all_feature = pd.merge(df_all_feature,df_model,on = 'app_order_id',how='left')   
    print(model,df_model.shape[0])
```

```python
df_all_feature
```

```python
#空值填充-999
df_all_feature.fillna(-999,inplace=True)
```

```python
df_all_feature = df_all_feature[df_all_feature['app_order_id'].isin(label['app_order_id'])]#确保df_all_feature的订单号在选取的样本内
```

```python
df_all_feature.index=range(df_all_feature.shape[0])  #重置索引
```

```python
import warnings
# 忽略所有警告
warnings.filterwarnings("ignore")
# 你的代码
```

```python
#删除所有数据类型不是数值型的列
for i in df_all_feature.columns[1:]:
    try:
        df_all_feature[i] = df_all_feature[i].astype(float)
    except:
        print(i)
        del df_all_feature[i]
```

```python
#保留所有单列值不是同一值的列
DF_columns = []
for i in df_all_feature.columns:
    if df_all_feature[i].value_counts().shape[0] > 1:
        DF_columns.append(i)
```

```python
#保存特征文件
df_all_feature[DF_columns].to_parquet('a11_all_feature.parquet')
```
